```python
import pandas as pd
import numpy as np
```

```python
credit_df = pd.read_csv('tmdb_5000_credits.csv')
movies_df = pd.read_csv('tmdb_5000_movies.csv')
```

### Note:
##### The column “movie_id” in credits_df table is same as the column “id” in movies_df table. Hence, this column will act as our bridge to connect these two tables.

```python
credit_column_renamed = credit_df.rename(index=str, columns={'movie_id':'id'})
```

```python
credit_column_renamed
```

```python
merged_df = movies_df.merge(credit_column_renamed, on='id')
```

```python
merged_df.head()
```

```python
merged_df.shape
```

```python
final_df = merged_df.drop(columns = ['homepage','title_x', 'title_y','status','production_countries'])
```

```python
final_df.info()
```

## Computing Weightage Average for Individual movie’s average rating

```python
final_df.head()
```

```python
v = final_df['vote_count']
R = final_df['vote_average']
C = final_df['vote_average'].mean()
m = final_df['vote_count'].quantile(0.70)
```

```python
v
```

```python
R
```

```python
C
```

```python
m
```

```python
final_df['weighted_average'] =((R*v + C*m)/(v+m))
```

```python
final_df.head()
```

```python
final_sorted_df=final_df.sort_values('weighted_average', ascending= False)
final_sorted_df
```

```python
weight_average = final_sorted_df[['original_title','popularity','vote_average','vote_count','weighted_average']]
```

```python
weight_average
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
plt.figure(figsize=(12,8))
axis1 = sns.barplot(x= weight_average['weighted_average'].head(10), y= weight_average['original_title'].head(10), data = weight_average)
plt.title('Best movies by average weight')
plt.xlim(4,10)
```

```python
popularity_df = weight_average.sort_values('popularity', ascending=False)
plt.figure(figsize=(14,10))
axis2 = sns.barplot(x=popularity_df['popularity'].head(10), y=popularity_df['original_title'].head(10), data= popularity_df)
```
